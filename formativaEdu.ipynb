```python
tel = []
for x in range(len(df['product_name'])):    
    if df['product_name'][x] == 'Televisão':        
        tel.append(df[x:(x+1)])
        

rating_tel = np.array([x['rating'] for x in tel])
compra_tel = np.array([y['purchased'] for y in tel])
print(rating_tel)
print(compra_tel)


cel = []
for x in range(len(df['product_name'])):    
    if df['product_name'][x] == 'Celular':        
        cel.append(df[x:(x+1)])
        

rating_cel = np.array([x['rating'] for x in cel])
compra_cel = np.array([y['purchased'] for y in cel])
print(rating_cel)
print(compra_cel)


caf = []
for x in range(len(df['product_name'])):    
    if df['product_name'][x] == 'Cafeteira':        
        caf.append(df[x:(x+1)])
        

rating_caf = np.array([x['rating'] for x in caf])
compra_caf = np.array([y['purchased'] for y in caf])
print(rating_caf)
print(compra_caf)


vidgame = []
for x in range(len(df['product_name'])):    
    if df['product_name'][x] == 'Videogame':        
        vidgame.append(df[x:(x+1)])
        

rating_vidgame = np.array([x['rating'] for x in vidgame])
compra_vidgame = np.array([y['purchased'] for y in vidgame])
print(rating_vidgame)
print(compra_vidgame)


iPod = []
for x in range(len(df['product_name'])):    
    if df['product_name'][x] == 'iPod':        
        iPod.append(df[x:(x+1)])
        

rating_iPod = np.array([x['rating'] for x in iPod])
compra_iPod = np.array([y['purchased'] for y in iPod])
print(rating_iPod)
print(compra_iPod)


Note = []
for x in range(len(df['product_name'])):    
    if df['product_name'][x] == 'Notebook':        
        Note.append(df[x:(x+1)])
        

rating_Note = np.array([x['rating'] for x in Note])
compra_Note = np.array([y['purchased'] for y in Note])
print(rating_Note)
print(compra_Note)


```

```python
rating_treino, rating_teste, compra_treino, compra_teste  = train_test_split(rating_tel.reshape(-1,1),compra_tel.reshape(-1,1),test_size =0.2,random_state=42)
rating_treino2, rating_teste2, compra_treino2, compra_teste2  = train_test_split(rating_cel.reshape(-1,1),compra_cel.reshape(-1,1),test_size =0.2,random_state=42)
rating_treino3, rating_teste3, compra_treino3, compra_teste3  = train_test_split(rating_caf.reshape(-1,1),compra_caf.reshape(-1,1),test_size =0.2,random_state=42)
rating_treino4, rating_teste4, compra_treino4, compra_teste4  = train_test_split(rating_vidgame.reshape(-1,1),compra_vidgame.reshape(-1,1),test_size =0.2,random_state=42)
rating_treino5, rating_teste5, compra_treino5, compra_teste5  = train_test_split(rating_iPod.reshape(-1,1),compra_iPod.reshape(-1,1),test_size =0.2,random_state=42)
rating_treino6, rating_teste6, compra_treino6, compra_teste6  = train_test_split(rating_Note.reshape(-1,1),compra_Note.reshape(-1,1),test_size =0.2,random_state=42)
```
